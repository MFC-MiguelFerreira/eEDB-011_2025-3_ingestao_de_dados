```python
import glob

import pandas as pd
from pandas.errors import EmptyDataError

pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

# Reading Data

## Enquadramento Inicial

```python
enq_inicial = pd.read_csv(
    "data/Bancos/EnquadramentoInicia_v2.tsv",
    sep="\t",
)
enq_inicial["Nome"] = enq_inicial["Nome"].str.strip().str.lower()
enq_inicial.info()
```

```python
enq_inicial["Nome"].value_counts(dropna=False)
```

```python
enq_inicial.head(15)
```

## Empregados

```python
empregados_csv_files = glob.glob(f"data/Empregados/*.csv")
list_of_dfs = []
for file in empregados_csv_files:
    df = pd.read_csv(file, sep="|")
    list_of_dfs.append(df)

empregados = pd.concat(list_of_dfs, ignore_index=True)
empregados.info()
```

```python
empregados.head(10)
```

## Reclamações

```python
reclamacoes_csv_files = glob.glob(f"data/Reclamacoes/*.csv")
list_of_dfs = []
for file in reclamacoes_csv_files:
    try:
        df = pd.read_csv(
            file,
            delimiter=';',
            encoding='latin1',
            decimal=',',
            skipinitialspace=True # Clean up leading spaces after delimiter
        )
        list_of_dfs.append(df)
    except EmptyDataError:
        pass

reclamacoes = pd.concat(list_of_dfs, ignore_index=True)
reclamacoes.info()
```

```python
reclamacoes.head(10)
```
